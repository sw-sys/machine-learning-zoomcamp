# Prepare and split the dataset
#### 1. Shuffle the dataset (the filtered one you created above), use seed 42.
#### 2. Split your data in train/val/test sets, with 60%/20%/20% distribution.

```python
n = len(df)

# do this to split
n_val = int(n * 0.2)
n_test = int(n * 0.2)
n_train = n - n_val - n_test

# test length
n, n_val + n_test + n_train
```

```python
df_val = df.iloc[:n_val]
df_test = df.iloc[n_val:n_val+n_test]
df_train = df.iloc[n_val+n_test:]
```

```python
# create index
idx = np.arange(n)
#declare a seed to make randomisation repeatable
np.random.seed(42)
# shuffle
np.random.shuffle(idx)
print(idx)
```

```python
#get rows for split sets via shuffled index

df_val = df.iloc[idx[:n_val]]
df_test = df.iloc[idx[n_val:n_val+n_test]]
df_train = df.iloc[idx[:n_train]]

len(df_train), len(df_val), len(df_test), 
```
